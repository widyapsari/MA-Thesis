```python
import pandas as pd
```

```python
## Basic data statistic ##

def generate_table(dataset_files):
    names = [] 
    instances = []
    not_offensive = [] 
    offensive = []
    offensive_percentage = []
    not_offensive_percentage=[]
    offensive_percentage_f=[]
    not_offensive_percentage_f=[]
    for file in dataset_files:
        names.append(file.split('.')[0])
        df = pd.read_csv(file)
        instances.append(len(df))
        not_offensive.append(df.value_counts("label").iloc[0])
        offensive.append(df.value_counts("label").iloc[1])
        offensive_percentage.append(int(df.value_counts("label").iloc[1]/len(df)*100))
        not_offensive_percentage.append(int(df.value_counts("label").iloc[0]/len(df)*100))
        offensive_percentage_f.append(df.value_counts("label").iloc[1]/len(df)*100)
        not_offensive_percentage_f.append(df.value_counts("label").iloc[0]/len(df)*100)
    df_stat = df = pd.DataFrame({
                    'Name':names,
                    'Instances': instances,
                    'Not_offensive': not_offensive,
                    'Offensive': offensive,
                    '%Offensive': offensive_percentage,
                    '%Not_Offensive': not_offensive_percentage,
                    '%Offensivef': offensive_percentage_f,
                    '%Not_Offensivef': not_offensive_percentage_f})
    
    return df_stat       
```

```python
## OLID ##

dataset_files = ["../data/OLID_train.csv", "../data/OLID_test.csv"]
generate_table(dataset_files) 
```

```python
## ALYT ##

dataset_files = ["../data/ALYT_train.csv", "../data/ALYT_test.csv"]
generate_table(dataset_files) 
```

```python
def compare_predictions_bert(path_df):
    df = pd.read_csv(path_df)
    df_standard = pd.read_csv('../predictions/alyt_obf_bert/ALYT_def_bert.csv')
    df.rename(columns = {'pred':'pred_obf'}, inplace = True)
    df['pred_standard'] = df_standard['pred'].tolist()
    df = df[(df['sw_found'] != '[]') & (df['label'] == 1)]
    return df
```
